# Fill Between Plot.

```python
import matplotlib.pyplot as plt
import numpy as np

# x=[1,2,3,4,5]
# y=[1,2,3,4,5]

x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.plot(x,y,color="b")

# plt.fill_between(x,y,color="b")
# plt.fill_between(x=[2,4],y1=2,y2=4,color="y")#x values 2 to 4
plt.fill_between(x,y,color="g",where=(x>=2) & (x<=4),alpha=0.3)#Range=2 to 4.

plt.title("Python")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```
