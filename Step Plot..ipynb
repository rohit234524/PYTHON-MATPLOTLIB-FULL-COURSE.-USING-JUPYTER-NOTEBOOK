# Step Plot.

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
#y=[1,2,3,4,5]

y=[3,2,3,1,5]
plt.step(x,y,marker="o",color="b",ms=10,mfc="g",label="R Language.")#Marker=point and Ms(markersize(pointsize))
# Marker Face Plot.

plt.title("R Language.")
plt.xlabel(" X-Axis ")
plt.ylabel(" Y-Axis")

plt.grid()
plt.legend(loc=1)# Label location Change.
plt.show()
```
