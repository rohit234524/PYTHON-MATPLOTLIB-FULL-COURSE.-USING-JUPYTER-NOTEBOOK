```python
import matplotlib.pyplot as plt
import numpy as np
x=["Python","C","C++","Java"] # X Parameter(1st Parameter)
y=[85,65,60,80] # Y Parameter(2nd parameter)
z=[10,20,30,40]

width=0.2
p=np.arange(len(x)) #Numpy array Create
p1=[j+width for j in p]

plt.xlabel("Language.",fontsize=15) # X-axis Data write.
plt.ylabel("Number",fontsize=15) # Y-axis Data write.
plt.title("Wscube Tech.",fontsize=20)

plt.bar(p,y,width=0.2,label="Popularity",color="b")
plt.bar(p1,z,width=0.2,label="Popularity1",color="y")
plt.xticks(p+width/2,x,rotation=15) #List-x data write(show).

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["Python","C","C++","Java"] # X Parameter(1st Parameter)
y=[85,65,60,80] # Y Parameter(2nd parameter)
z=[10,20,30,40]

width=0.2
p=np.arange(len(x)) #Numpy array Create
p1=[j+width for j in p]

plt.xlabel("Language.",fontsize=15) # X-axis Data write.
plt.ylabel("Number",fontsize=15) # Y-axis Data write.
plt.title("Wscube Tech.",fontsize=20)

plt.barh(p,y,label="Popularity",color="b")#Horizontal Bar Chart.
plt.bar(p1,z,label="Popularity1",color="y")#Vertical Bar chart
plt.xticks(p+width/2,x,rotation=15) #List-x data write(show).

plt.legend()
plt.show()
```
