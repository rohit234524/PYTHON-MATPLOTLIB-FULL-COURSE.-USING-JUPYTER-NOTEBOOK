# Stem Plot(Dot and Linear Plot).

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,2,5,4,6]

# plt.stem(x,y)
# plt.stem(x,y,linefmt=":",markerfmt="r+")# linefmt pamameter(line shape change). 
#plt.stem(x,y,linefmt=":",markerfmt="r-")#markerfmt parameter(marker(dot) shape chage with color.)

plt.stem(x,y,linefmt=":",markerfmt="m*",bottom=0)# Bottom Parameter(bottom line change).
plt.stem(x,y,linefmt=":",basefmt="b",label="Python data",use_line_collection=False,orientation="horizontal")#Basefmt parameter(base color change).
#Use_line_collection parmeter(Base line with color change label).
plt.stem(x,y,linefmt=":",basefmt="b",orientation="horizontal")


plt.legend()
plt.show()
```
