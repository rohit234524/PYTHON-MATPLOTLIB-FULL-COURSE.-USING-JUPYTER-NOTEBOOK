# Pie Plot.

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.3,0.0,0.0,0.0]

c=["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1.2,
        labeldistance=1,startangle=0,textprops={"fontsize":34},counterclock=False,
        wedgeprops={"linewidth":6,"width":2,"edgecolor":"r"},center=(2,4),
        rotatelabels=False)#Startangle=Start point to rotate Pie Plot.#cunterclock=counter by clock rotation pie chart.

plt.title("Wscube Tech.",fontsize=25)
plt.legend(loc=2)
plt.show()
```
