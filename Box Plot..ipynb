# Box Plot.

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70,130]

# plt.boxplot(x,notch=True) #North Box-Plot.
# plt.boxplot(x,vert=False,notch=True) #Horizontal Box-Plot.
# plt.boxplot(x,vert=True,widths=0.4,labels=["Python"]) #Bi-default width=0.2.#labels value store in list(because multiple boxplot Create).
 
# plt.boxplot(x,vert=True,patch_artist=True,showmeans=True)#Patch_artist(color fill in boxplot).
#showmeans mean value point(show). 
# plt.boxplot(x,vert=True,showmeans=True,whis=1.2,sym="b+")#whisk join to outlier use whis parameter.
#sym parameter(outlier shape change with color). 
    
plt.boxplot(x,sym="g+",boxprops=dict(color="r"),capprops=dict(color="g"),whiskerprops=dict(color="b"),flierprops=dict(markeredgecolor="y")) # boxprops parameter(boxplot color change)    
#Capprops( cap color Change).
#Flierprops (fliercolor change).


plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70,130]
x1=[10,20,30,40,50,60,70]

y=[x,x1]

plt.boxplot(y,labels=["Python","Java"],sym="g+",boxprops=dict(color="r"),capprops=dict(color="g"),whiskerprops=dict(color="b"),flierprops=dict(markeredgecolor="y"),showmeans=True)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]
x1=[10,20,30,40,50,60,70]

y=[x,x1]

plt.boxplot(x,labels=["Python"],sym="g+",boxprops=dict(color="r"),capprops=dict(color="g"),whiskerprops=dict(color="b"),flierprops=dict(markeredgecolor="y"),showmeans=True)
plt.boxplot(x1,labels=["Python"],sym="g+",boxprops=dict(color="r"),capprops=dict(color="g"),whiskerprops=dict(color="b"),flierprops=dict(markeredgecolor="y"),showmeans=True)

plt.show()
```
