# Working With Text in MatPlotlib Python.

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.title("WsCube",fontsize=15)#title=Kiske information se related kowledge de raha hai ye graph.
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)


plt.text(2,5,"Rohit",fontsize=20,style="italic",bbox={"facecolor":"b"})

# plt.annotate("python",xy=(2,1),xytext=(3,3),arrowprops=dict(facecolor="black",shrink=100)
plt.plot(x,y)
plt.legend(["UP"],loc=9,facecolor="Red",edgecolor="yellow")# Loc Values=0-10
plt.show()
```
