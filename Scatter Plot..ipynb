# Scatter Plot.

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6]
num=[2,5,2,1,7,8]
num1=[3,6,7,9,2,5]

size=[500,200,300,600,450,120]

#colors=["r","y","b","m","g","r"]
colors=[10,20,49,55,67,54]

#plt.scatter(day,num,color="r")
#plt.scatter(day,num,c=colors,s=size)# s Parameter=scatter size
#plt.scatter(day,num,c=colors,s=size,alpha=1,marker="*",edgecolor="b",linewidth=2)# alpha=bolor line transparancy(value=0-1).

plt.scatter(day,num,c=colors,s=size,cmap="viridis") # Multiple color use color Function.
plt.scatter(day,num1,color="r",s=size,alpha=1)

plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)
plt.title("Scatter Plot",fontsize=20)

t=plt.colorbar()
t.set_label("ColorBar.",fontsize=15)
plt.show()
```
