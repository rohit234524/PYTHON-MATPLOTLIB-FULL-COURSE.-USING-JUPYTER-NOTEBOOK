# X-ticks(),Y-Ticks(),xlim(),ylim() axis() in Matplotlib Python.

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[3,1,5,2,4]

plt.plot(x,y)

# plt.xticks(x)
# plt.yticks(x)
plt.xticks(x,labels=["Python","c","C++","Java","R"])

plt.xlim(0,10)#X-axis Limit Increasing.
plt.ylim(0,10)#Y-axis limit(Data) Increasing.
plt.axis([0,10,0,7])

plt.show()
```
