# Sub Plot.

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.subplot(2,2,1)
plt.plot(x,y,color="r")
plt.title("R")

plt.subplot(2,2,2)
plt.title("C++")
plt.xlabel("Data.")
x1=[10,20,30,40]
plt.pie([1],colors="b")# One Radius Pie Plot Draw.

plt.subplot(2,2,3)
plt.pie(x)
plt.title("Python.")

x2=["a","s","d","f"]
y2=[10,20,30,40]
plt.subplot(2,2,4)
plt.bar(x2,y2,color="b")
plt.title("Java Programmer.")

plt.show()
```
